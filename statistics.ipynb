## Статистические тесты

Сгенерировать 10000 значений из нормального распределения (среднее - 10, дисперсия - 3). Сохранить в падасовскую серию

```python
s = pd.Series(sps.norm.rvs(loc=10, scale=3, size=10000))
```

Это будет наша генеральная совокупность. Посчитаем ее среднее

```python
p = s.mean()
```

Вытащим из нашей гс выборку из 50 элементов и сохраним ее

```python
q1 = s.sample(50)
```

Посчитаем среднее нашей выборки

```python
q1.mean()
```

Вытащим из нашей гс вторую выборку из 50 элементов и сохраним ее

```python
q2 = s.sample(50)
```

Посчитаем среднее второй выборки

```python
q2.mean()
```

Если бы мы не знали, что обе выборки из одной гс, то как мы бы ответили на вопрос: из одной ли гс обе выборки?
Применили бы статистический тест. Какая нулевая гипотеза?

Начнем с t-теста Стьюдента

```python
sps.ttest_ind(q1, q2)[1]
```

Отбрасываем нулевую гипотезу только при pvalue < 0.05

Проведем такие симуляции 10 000 раз, сохраняя полученные значения p-value в список

```python
q = []
for a in range(10000):
    q.append(sps.ttest_ind(s.sample(50), s.sample(50))[1])
```

Как будет выглядеть гистограмма p-значений?

Построить гистограмму этого списка

```python
sns.histplot(q, bins = 100)
```

Как часто наш тест будет ошибаться?

```python
q = pd.Series(q)
```

```python
q[q < 0.05].count()
```

Сделать то же самое с непараметрическим тестом Манна-Уитни

```python
q = []
for a in range(10000):
    q.append(sps.mannwhitneyu(s.sample(50), s.sample(50))[1])
```

Построить гистограмму этого списка

```python
sns.histplot(q)
```
